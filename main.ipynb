```python
#dependencies 
import pandas as pd
```

```python
# import csv files
student_file = 'students_complete.csv'
schools_file = 'schools_complete.csv'

student_data_df = pd.DataFrame(pd.read_csv(student_file, encoding='iso-8859-1', low_memory=False))
school_data_df = pd.DataFrame(pd.read_csv(schools_file, encoding='iso-8859-1', low_memory=False))
school_data_df.head()
#school_data_df
```

```python
student_data_df.head()
```

```python
#rename columns to allow for data to be merged

school_data_df.columns = ["School ID", "School Name", "type","size","budget"]
student_data_df.columns = ["Student ID", "Student Name", "gender","grade","School Name", "reading_score", "math_score"]

student_data_df.head()
```

```python
#merge dfs 

merge_data = pd.merge(school_data_df, student_data_df, on="School Name")

merge_data.head()
```

```python
total_schools = len(merge_data["School Name"].unique())
total_schools
```

```python
total_students = len(merge_data["Student Name"].unique())
total_students
```

```python
total_budget = sum(school_data_df["budget"])
total_budget
```

```python
average_math_score = merge_data["math_score"].mean()
average_math_score
```

```python
average_reading_score = merge_data["reading_score"].mean()
average_reading_score
```

```python

passing_math = merge_data.loc[merge_data["math_score"] > 64,:]
pct_pass_math = len(passing_math) / len(merge_data)
pct_pass_math
```

```python
passing_read = merge_data.loc[merge_data["reading_score"] > 64,:]
pct_pass_read = len(passing_read) / len(merge_data)
pct_pass_read
```

```python
overall_pass = (pct_pass_math + pct_pass_read) / 2
overall_pass
```

```python
#pd.DataFrame({'A' : []})

dist_summary = pd.DataFrame({"% Passing Math": pct_pass_math, "% Passing Reading": pct_pass_read, "% Overall Passing":overall_pass,
                             'Total Schools': [total_schools], 'Total Students': [total_students],
                             'Total Budget': [total_budget],"Average Math Score":average_math_score, "Average Reading Score": average_reading_score,})

dist_summary = dist_summary[['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score',
                       '% Passing Math','% Passing Reading','% Overall Passing']]
dist_summary
```
